```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('ex1.csv')
data1.dropna(inplace=True)
data1.head()
```

```python
# Boolean mask
data1_bm = data1['A'] > 0.5
data1_bm
```

```python
# only data which met the condition was retained, otherwise its NaN
data1.where(data1_bm)
```

```python
# If we don't want the NaN data, we use the dropna() function
data1.where(data1_bm).dropna()
```

```python
# Or we can just do this
data1[data1['A'] > 0.5]
```

```python
# Project a single column
data1['A']
```

```python
# Project a list of columns
data1[['A', 'B']]
```

```python
# Combining multiple boolean masks
(data1['A'] > 0.5) & (data1['A'] < 0.6)
```

```python
# Another one way to do this is by using the built-in function
data1['A'].gt(0.5) & data1['A'].lt(0.6)
```

```python
# You can chain them together too
data1['A'].gt(0.5).lt(0.6)
```
