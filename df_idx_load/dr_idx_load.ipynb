```python
import pandas as pd
```

```python
# turn csv into a dataframe
df = pd.read_csv('ex1.csv')

# look at the few first rows
df.head()
```

```python
# Set the index by using 'index_col'
df = pd.read_csv('ex1.csv', index_col=0)
df.head()
```

```python
# change the column names by using the rename() function
new_df = df.rename(columns={'A ': 'Apple',
                           'B': 'Banana',
                           'C': 'Coconut'})
new_df.head()
```

```python
# As you can see 'A' didn't change. Let's investigate
new_df.columns
```

```python
# if we look at the output closely, there's a space right after the 'A'
# let's strip the whitespace using "strip()"
new_df = new_df.rename(mapper=str.strip, axis='columns')
new_df.head()
```

```python
# change all the column names to lower case
cols = list(new_df)
cols = [i.lower().strip()for i in cols]
new_df.columns = cols
new_df.head()
```
